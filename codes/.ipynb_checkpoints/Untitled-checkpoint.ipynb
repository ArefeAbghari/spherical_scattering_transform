```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.io.fits as fits
import pandas as pd
import healpy as hp
import time
import warnings
from pixell import reproject
from classy import Class
%matplotlib inline
import ST
import pysm3
```

```python
map_nosz = hp.read_map("/Users/arefe/Projects/ScatteringTransform/maps/COM_CMB_IQU-smica-nosz_2048_R3.00_full.fits")
```

```python
map_nosz_0256 = hp.ud_grade(map_nosz , 256)
```

```python
nside_0256 = 256 
lmax_0256 = 3*nside_0256-1 
ell_0256=np.arange(0,lmax_0256+1)
norm_0256 = ell_0256 * (ell_0256+1) / (2*np.pi)
ellsc_0256 = 360*60/ell_0256
resol_0256 = hp.nside2resol(nside_0256 , arcmin = False )
resol_0256_arcmin = hp.nside2resol(nside_0256 , arcmin=True)
```

```python
jmax = 7
theta_bin = 10000
Jf=np.arange(jmax)
scale_0256=2**Jf*((2*resol_0256_arcmin)/0.75)
```

```python
mor_l_0256 = ST.morlet_arr (resol_0256, jmax , lmax_0256, theta_bin)
gaus_l_0256 = ST.gaus_arr (resol_0256, jmax, lmax_0256, theta_bin)
```

```python
s1_nosz_0256 , i1_nosz_0256 = ST.compS1 (map_nosz_0256 , mor_l_0256, jmax, nside_0256 )
```

```python
%%time 
s2_nosz_0256 , i2_nosz_0256 = ST.compS2 ( i1_nosz_0256, mor_l_0256, jmax,  nside_0256 )
```

```python
cl_nosz_0256 = hp.anafast(map_nosz_0256)
dl_nosz_0256 = cl_nosz_0256*norm_0256
```

```python
plt.plot(ellsc_0256,np.log(dl_nosz_0256))
plt.xlabel("scale")
plt.ylabel("$D_l$")
plt.xlim (0,2000)
```

```python
plt.figure(figsize=(20,10))
plt.plot (scale_0256,np.log(s1_nosz_0256),'o-')
plt.xlabel("scale(arcmin)")
plt.ylabel("log(S1)")
```

```python
fig, ax = plt.subplots(1, jmax,sharey=True,squeeze=True,figsize=(25,10))
for j in range (jmax):
    ax[j].plot (scale_0256[j+1:],np.log(s2_nosz_0256[j,j+1:]/s1_nosz_0256[j]),'o-' , alpha =1 , color = "blue")
    #ax[j].plot (scale[j+1:],np.log(s2_sz_out[j,j+1:]/s1_sz_out[j]),'o-' , alpha = 1 , color = "red")
    #ax[j].plot (scale,np.log(s2_sz_out[j,:]/s1_sz_out[j]),'o-' , alpha = 0.4 , color = "red")
    ax[j].plot (scale_0256,np.log(s2_nosz_0256[j,:]/s1_nosz_0256[j]),'o-' , alpha =0.4 , color = "blue")
    ax[j].set_title("j1=%d"%j)
    ax[j].set_xscale('log')
    ax[j].legend()
#fig.legend(['SZ map' ,'Regenerated SZ map'])
fig.suptitle("S2 coefficients for NoSZ map nside=256")
```
