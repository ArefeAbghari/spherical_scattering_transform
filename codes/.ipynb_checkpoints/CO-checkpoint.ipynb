```python
map_co = hp.read_map ("/Users/arefe/Projects/ScatteringTransform/maps/COM_CompMap_CO-commander_0256_R2.00.fits")
```

```python

hp.mollview(map_co  ,title="CO map")
```

```python
nside = hp.pixelfunc.get_nside(map_co)
print (nside)
```

```python
res_co = hp.nside2resol (nside, arcmin= False )
res_co_arcmin = hp.nside2resol(nside , arcmin= True)
```

```python
cl_co = hp.anafast(map_co)
```

```python
lmax=np.shape(cl_co)[0]
ell=np.arange(0,lmax)
norm = ell * (ell+1) / (2*np.pi)
dl_co = cl_co * norm
```

```python
ellsc = 360*60/ell
plt.plot(ellsc,np.log(dl_co))
plt.xlabel("scale")
plt.ylabel("$log(D_l)$")
```

```python
alm_co = hp.map2alm (map_co)
```

```python
resol = hp.nside2resol(nside, arcmin=False )
resol_arcmin = hp.nside2resol(nside, arcmin=True )
print (resol)
```

```python
jmax = 10
theta_bin = 10000
```

```python
mor_l = s1.morlet_arr (resol, jmax , lmax, theta_bin)
gaus_l = s1.gaus_arr (resol, jmax, lmax, theta_bin)
```

```python
s1_co , i1_co = s1.compS1 (map_co , mor_l, gaus_l, resol, jmax , lmax , nside)
```

```python
Jf=np.arange(jmax)
res256_arcmin = hp.nside2resol(256, arcmin=True)
scale=2**Jf*((2*res256_arcmin)/0.75)
plt.figure(figsize=(20,10))
plt.plot (scale,np.log(s1_co),'o-')
plt.xlabel("scale(arcmin)")
plt.ylabel("log(S1)") 
```

```python
s2_co  = s1.compS2 (map_co , i1_co, mor_l, gaus_l, resol, jmax , lmax , nside)
```

```python
fig, ax = plt.subplots(1, jmax,sharey=True,squeeze=True,figsize=(25,10))
for j in range (jmax):
    ax[j].plot (scale,np.log(s2_co[j,:]/s1_co[j]),'o-')
    ax[j].set_title("j1=%d"%j)
    ax[j].set_xscale('log')
    ax[j].legend()
```
