```python
sin_theta_array = np.sin(theta_array)
cos_theta_array = np.cos(theta_array)
```

```python
sin_theta_2d_array = np.outer(sin_theta_array, sin_theta_array)
cos_theta_2d_array = np.outer(cos_theta_array, cos_theta_array)
delta_phi_array = np.cos(phi_array - phi_array[:,None])
```

```python
distance_array = np.arccos(cos_theta_2d_array + sin_theta_2d_array * delta_phi_array)
```

```python
def gaussian_window (sigma):
    def window (r):
        return (np.exp(-(r**2)/2/sigma**2)/sigma**2)
    return window
```

```python
quasar_map = hp.read_map('/Users/arefe/Downloads/quasar_map-1.fits')
quasar_map = hp.ud_grade(quasar_map,nside)
```

```python
quasar_mask = hp.read_map('/Users/arefe/Downloads/quasar_mask.fits')
quasar_mask = hp.ud_grade(quasar_mask,nside)
#quasar_mask = np.logical_not(quasar_mask)
```

```python
quasar_masked_map = hp.ma (quasar_map)
quasar_masked_map.mask = np.logical_not(quasar_mask)
hp.mollview (quasar_masked_map)
```

```python
quasar_zerod_map = quasar_map*quasar_mask
hp.mollview(quasar_zerod_map)
```

```python
quasar_zerod_map[quasar_mask==0] = np.nan
```

```python
smoothed_quasar_masked_map = np.zeros(map_size)
```

```python
for i in range (map_size):
    if (np.isnan(quasar_zerod_map[i])==False):
        #print (i)
        weights = gaussian_window(0.02)(distance_array[i,:])
        smoothed_quasar_masked_map [i] = np.nansum (weights*quasar_zerod_map)
```
